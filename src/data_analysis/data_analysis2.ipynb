```python
import pandas as pd
import re

def parse_numeric_string(value):
    """
    Extracts digits from the given value and returns them as a float.
    - Returns None if the input is NaN or contains no digits.
    - If the input is already a number, it’s returned as a float.
    """
    if pd.isna(value):
        return None


    if isinstance(value, (int, float)):
        return float(value)

    text = str(value)
    digit_str = re.sub(r'\D+', "", text)

    return float(digit_str) if digit_str else None
```

```python
import os

# 1. Define which columns need numeric extraction
columns_to_process = ['Mileage', 'Engine', 'Power', 'New_Price']

# 2. Apply the parsing function to each column
for col in columns_to_process:
    df[col] = df[col].apply(parse_numeric_string)

# 3. Ensure the output directory exists
cleaned_dir = "/content/cleaned_dataset"
os.makedirs(cleaned_dir, exist_ok=True)

# 4. Save the cleaned DataFrame
output_path = os.path.join(cleaned_dir, "train.csv")
df.to_csv(output_path, index=False)

print(f"Cleaned data saved to: {output_path}")
```

```python
import os

# Ensure the directory for cleaned data exists
clean_dir = "/content/cleaned_dataset"
os.makedirs(clean_dir, exist_ok=True)

# 1. Pick the core columns we care about
subset_df = df.loc[:, ['Name', 'Year', 'Fuel_Type', 'Transmission', 'Price']]

# 2. Keep only vehicles built after 2016
recent_models = df[df['Year'] > 2016]

# 3. Rename a couple of columns for clarity
rename_map = {'Mileage': 'Efficiency', 'Power': 'Engine'}
renamed_df = df.rename(columns=rename_map)

# 4. Add a column for the car’s age (as of 2024)
df['Vehicle_Age'] = 2024 - df['Year']

# 5. Sort all entries by Price, from highest to lowest
ranked_by_price = df.sort_values(by='Price', ascending=False)

# 6. Compute the average Price per Fuel_Type
avg_price = df.groupby('Fuel_Type')['Price'].mean()

# 7. Compute mean, min, and max Price per Fuel_Type
detailed_price_stats = df.groupby('Fuel_Type')['Price'].agg(['mean', 'min', 'max'])

# Print out results
print("=== Subset of Key Columns ===")
print(subset_df.head(), "\n")

print("=== Models After 2016 ===")
print(recent_models.head(), "\n")

print("=== After Renaming Columns ===")
print(renamed_df.head(), "\n")

print("=== Vehicle Age Column ===")
print(df[['Name', 'Year', 'Vehicle_Age']].head(), "\n")

print("=== Top-Priced Vehicles ===")
print(ranked_by_price.head(), "\n")

print("=== Average Price by Fuel Type ===")
print(avg_price, "\n")

print("=== Detailed Price Stats by Fuel Type ===")
print(detailed_price_stats, "\n")

# Save the final cleaned DataFrame
output_path = os.path.join(clean_dir, "train.csv")
df.to_csv(output_path, index=False)
print(f"Cleaned data saved to: {output_path}")
```
