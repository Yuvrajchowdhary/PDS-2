```python
import pandas as pd
```

```python
df = pd.read_csv("/content/train.csv")
df
```

```python
null_counts = df.isna().sum()

print("Missing‑Value Summary per Column:\n")

for column_name, missing_count in null_counts.items():
    print(f"{column_name}: {missing_count}")
```

```python
null_summary = df.isna().sum()


cols_to_fill = null_summary[null_summary > 0].index


for col in cols_to_fill:

    pct_missing = (df[col].isna().sum() / len(df)) * 100
    print(f"Filling {pct_missing:.2f}% missing in '{col}'")

    if pd.api.types.is_numeric_dtype(df[col]):
        impute_val = df[col].mean()
    else:
        impute_val = df[col].mode().iloc[0]

    df[col] = df[col].fillna(impute_val)
```

```python
df.isnull().sum()
```

```python
import os


cleaned_dir = "/content/cleaned_dataset"
os.makedirs(cleaned_dir, exist_ok=True)

output_path = os.path.join(cleaned_dir, "cleaned_train.csv")

df.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to: {output_path}")
```
